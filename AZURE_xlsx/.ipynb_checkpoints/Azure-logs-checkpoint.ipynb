```python
import pandas as pd

AWS_js_hello = "AWS-js-hello.csv"
AWS_java_hello = "AWS-java-hello.csv"
AWS_go_hello = "AWS-go-hello.csv"

AWS_js_dynamo = "AWS-js-dynamo.csv"
AWS_java_dynamo = "AWS-java-dynamo.csv"
AWS_go_dynamo = "AWS-go-dynamo.csv"


df = pd.read_csv(AWS_go_dynamo)
df.sort_values(by='timestamp', ascending=True)
```

```python
list1 = []
list1 = df["message"].tolist()

list2 = []
for x in list1:
    list2.append(x.split(" "))

list3 = []
for x in list2:
    list3.append(int(x[-2]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array(list3)

fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Duration of service retrieval from DynamoDB")
ax.set_xlabel('Invocations')
ax.set_ylabel('Duration [ms]')
ax.plot(y)
ax.grid(True, linewidth=0.5)
ax.set_yticks(range(0,1550,50))
plt.show() 

```
