```python
import pandas as pd

Azure_js_hello = "Azure-js-hello.csv"
Azure_java_hello = "Azure-java-hello.csv"
Azure_go_hello = "Azure-go-hello.csv"

Azure_js_cosmos = "Azure-js-cosmos.csv"
Azure_java_cosmos = "Azure-java-cosmos.csv"
Azure_go_cosmos = "Azure-go-cosmos.csv"


df = pd.read_csv(Azure_js_cosmos).sort_values(by='timestamp [UTC]', ascending=True)
df1 = pd.read_csv(Azure_java_cosmos).sort_values(by='timestamp [UTC]', ascending=True)
df2 = pd.read_csv(Azure_go_cosmos).sort_values(by='timestamp [UTC]', ascending=True)
df1.drop(df1.index[:5], inplace=True)
df1.head(20)
```

```python
import numpy as np


list1 = []
list1 = df["message"].tolist()
list2 = []
for x in list1:
    list2.append(x.split(" "))
list3 = []
for x in list2:
    list3.append(int(x[-2]))
y_js = np.array(list3)


list1 = []
list1 = df1["message"].tolist()
list2 = []
for x in list1:
    list2.append(x.split(" "))
list3 = []
for x in list2:
    list3.append(int(x[-2]))
y_java = np.array(list3)


list1 = []
list1 = df2["message"].tolist()
list2 = []
for x in list1:
    list2.append(x.split(" "))
list3 = []
for x in list2:
    list3.append(int(x[-2]))
y_go = np.array(list3)
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Duration of invocation and execution of the hello-world Go function", fontsize = 40)
ax.set_xlabel('Invocations', fontsize = 34)
ax.set_ylabel('Duration [ms]', fontsize = 34)
ax.plot(y_js, label = 'js function')
ax.plot(y_java, label = 'java function')
ax.plot(y_go, label = 'go function')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
ax.grid(True, linewidth=0.5)
plt.yscale ('log')
#ax.set_yticks(range(0,2200,100))
plt.legend(loc = 'upper right')
plt.rc('legend', fontsize = 25)
plt.show() 
```
