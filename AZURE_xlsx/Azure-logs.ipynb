```python
import pandas as pd

Azure_js_hello = "Azure-js-hello.csv"
Azure_java_hello = "Azure-java-hello.csv"
Azure_go_hello = "Azure-go-hello.csv"

Azure_js_cosmos = "Azure-js-cosmos.csv"
Azure_java_cosmos = "Azure-java-cosmos.csv"
Azure_go_cosmos = "Azure-go-cosmos.csv"


df = pd.read_csv(Azure_go_cosmos)
df = df.sort_values(by='timestamp [UTC]', ascending=True)
#df.drop(df.index[:5], inplace=True)
df.head(10)
```

```python
list1 = []
list1 = df["message"].tolist()

list2 = []
for x in list1:
    list2.append(x.split(" "))

list3 = []
for x in list2:
    list3.append(int(x[-2]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array(list3)

fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Duration of service retrieval from CosmosDB")
ax.set_xlabel('Invocations')
ax.set_ylabel('Duration [ms]')
ax.plot(y)
ax.grid(True, linewidth=0.5)
ax.set_yticks(range(0,300,50))
plt.show() 

```
