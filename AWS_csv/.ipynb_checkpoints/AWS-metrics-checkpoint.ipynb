```python
import pandas as pd

AWS_js_duration = "AWS-js-duration.csv"
AWS_js_invocation = "AWS-js-invocation.csv"
AWS_js_throttles = "AWS-js-throttles.csv"

AWS_java_duration = "AWS-java-duration.csv"
AWS_java_invocation = "AWS-java-invocation.csv"
AWS_java_throttles = "AWS-java-throttles.csv"

AWS_go_duration = "AWS-go-duration.csv"
AWS_go_invocation = "AWS-go-invocation.csv"
AWS_go_throttles = "AWS-go-throttles.csv"

df = pd.read_csv(AWS_js_invocation)
df1 = pd.read_csv(AWS_java_invocation)
df2 = pd.read_csv(AWS_go_invocation)
```

```python
#plotting the durations

import matplotlib.pyplot as plt
import numpy as np

'''list1 = df1.iloc[4:,0].tolist()

list2 = []
for x in list1:
    list2.append(x.split(" "))

list3 = []
for x in list2:
    list3.append(x[-1])'''

list3 = [0, 60, 120, 180, 240, 300, 360]
x = np.array(list3)

y = np.array (df.iloc[4:, 1].tolist())
y_min = [0]
for item in y:
    y_min.append(float(item))
y_min.append(0)
y = np.array (df.iloc[4:, 2].tolist())
y_avg = [0]
for item in y:
    y_avg.append(float(item))
y_avg.append(0)
y = np.array (df.iloc[4:, 3].tolist())
y_max = [0]
for item in y:
    y_max.append(float(item))
y_max.append(0)


y1 = np.array (df1.iloc[4:, 1].tolist())
y_min1 = []
for item in y1:
    y_min1.append(float(item))
y1 = np.array (df1.iloc[4:, 2].tolist())
y_avg1 = []
for item in y1:
    y_avg1.append(float(item))
y1 = np.array (df1.iloc[4:, 3].tolist())
y_max1 = []
for item in y1:
    y_max1.append(float(item))


y2 = np.array (df2.iloc[4:, 1].tolist())
y_min2 = []
for item in y2:
    y_min2.append(float(item))
y_min2.append(0)
y2 = np.array (df2.iloc[4:, 2].tolist())
y_avg2 = []
for item in y2:
    y_avg2.append(float(item))
y_avg2.append(0)
y2 = np.array (df2.iloc[4:, 3].tolist())
y_max2 = []
for item in y2:
    y_max2.append(float(item))
y_max2.append(0)



fig, (ax1, ax2, ax3) = plt.subplots(3, figsize = (12,10)) 
ax1.set_title("Min duration of execution", fontsize = 14)
#axis[0,0].set_xlabel('Time', fontsize = 14)
ax1.set_ylabel('Duration [ms]', fontsize = 14)

ax2.set_title("Avg duration of execution", fontsize = 14)
#axis[0,1].set_xlabel('Time', fontsize = 14)
ax2.set_ylabel('Duration [ms]', fontsize = 14)

ax3.set_title("Max duration of execution", fontsize = 14)
#axis[1,0].set_xlabel('Time', fontsize = 14)
ax3.set_ylabel('Duration [ms]', fontsize = 14)

ax1.plot(x, y_min, label = "js function")
ax2.plot(x, y_avg, label = "js function")
ax3.plot(x, y_max, label = "js function")
ax1.plot(x, y_min1, label = "java function")
ax2.plot(x, y_avg1, label = "java function")
ax3.plot(x, y_max1, label = "java function")
ax1.plot(x, y_min2, label = "go function")
ax2.plot(x, y_avg2, label = "go function")
ax3.plot(x, y_max2, label = "go function")
plt.rc('legend', fontsize = 10)
ax1.grid(True, linewidth=0.5)
ax2.grid(True, linewidth=0.5)
ax3.grid(True, linewidth=0.5)
ax1.legend(loc = 'upper right')
ax2.legend(loc = 'upper right')
ax3.legend(loc = 'upper right')
ax1.set_xticks(range(0,420,60))
ax2.set_xticks(range(0,420,60))
ax3.set_xticks(range(0,420,60))
plt.show()
```

```python
#plotting the invocations

import matplotlib.pyplot as plt
import numpy as np

'''list1 = df1.iloc[4:,0].tolist()

list2 = []
for x in list1:
    list2.append(x.split(" "))

list3 = []
for x in list2:
    list3.append(x[-1])'''

list3 = [0, 60, 120, 180, 240, 300, 360]
x = np.array(list3)


y = np.array (df.iloc[4:, 1].tolist())
invocations = [0]
for item in y:
    invocations.append(int(item))
invocations.append(0)

y1 = np.array (df1.iloc[4:, 1].tolist())
invocations1 = []
for item in y1:
    invocations1.append(int(item))



y2 = np.array (df2.iloc[4:, 1].tolist())
invocations2 = []
for item in y2:
    invocations2.append(int(item))
invocations2.append(0)

fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Number of invocations per minute", fontsize = 40)
ax.set_xlabel('Seconds since beginning', fontsize = 34)
ax.set_ylabel('Number of invocations', fontsize = 34)
ax.plot( x, invocations, label = "javascript function")
ax.plot( x, invocations1, label = "java function")
ax.plot(x, invocations2, label = "go function")
plt.xticks(fontsize=30)
ax.set_xticks(range(0,420,60))
plt.yticks(fontsize=30)
plt.rc('legend', fontsize = 25)
plt.legend(loc = 'upper right')
ax.grid(True, linewidth=0.5)
plt.show()
```
