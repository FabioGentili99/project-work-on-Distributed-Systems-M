```python
import pandas as pd

Azure_js_metrics = "Azure-js-metrics.xlsx"
Azure_java_metrics = "Azure-java-metrics.xlsx"
Azure_go_metrics = "Azure-go-metrics.xlsx"


xls = pd.read_excel(Azure_js_metrics, sheet_name=['Somma Requests per DDI-first-ar', 'Media Response Time per DDI-fir', 'Min Response Time per DDI-first', 'Max Response Time per DDI-first'])
df1 = xls['Somma Requests per DDI-first-ar']
df2 = xls['Media Response Time per DDI-fir']
df3 = xls['Min Response Time per DDI-first']
df4 = xls['Max Response Time per DDI-first']
df1.drop(df1.index[:10], inplace=True)
df2.drop(df2.index[:10], inplace=True)
df3.drop(df3.index[:10], inplace=True)
df4.drop(df4.index[:10], inplace=True)
```

```python
#plotting the durations

import matplotlib.pyplot as plt
import numpy as np

list1 = df2.iloc[:,0].tolist()

list2 = []
for x in list1:
    #list2.append(x.split(" "))
    list2.append(x.date())

#list3 = []
#for x in list2:
#    list3.append(x[-1])

x = np.array(list2)

y = np.array (df3.iloc[:, 1].tolist())
y_min = []
for item in y:
    y_min.append(float(item))
y = np.array (df2.iloc[:, 1].tolist())
y_avg = [0]
for item in y:
    y_avg.append(float(item))
y_avg.append(0)
y_avg.append(0)
y = np.array (df4.iloc[:, 1].tolist())
y_max = []
for item in y:
    y_max.append(float(item))



fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Duration of execution (js function)")
ax.set_xlabel('Time')
ax.set_ylabel('Duration [ms]')
ax.plot(x, y_min, label = "minimum duration")
ax.plot(x, y_avg, label = "average duration")
ax.plot(x, y_max, label = "maximum duration")
ax.grid(True, linewidth=0.5)
plt.legend()
plt.show()
```
