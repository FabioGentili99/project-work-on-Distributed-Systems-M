```python
import pandas as pd

AWS_js_duration = "AWS-js-duration.csv"
AWS_js_invocation = "AWS-js-invocation.csv"
AWS_js_throttles = "AWS-js-throttles.csv"

AWS_java_duration = "AWS-java-duration.csv"
AWS_java_invocation = "AWS-java-invocation.csv"
AWS_java_throttles = "AWS-java-throttles.csv"

AWS_go_duration = "AWS-go-duration.csv"
AWS_go_invocation = "AWS-go-invocation.csv"
AWS_go_throttles = "AWS-go-throttles.csv"

df = pd.read_csv(AWS_go_throttles)
```

```python
#plotting the durations

import matplotlib.pyplot as plt
import numpy as np

list1 = df.iloc[4:,0].tolist()

list2 = []
for x in list1:
    list2.append(x.split(" "))

list3 = []
for x in list2:
    list3.append(x[-1])

x = np.array(list3)

y = np.array (df.iloc[4:, 1].tolist())
y_min = []
for item in y:
    y_min.append(float(item))
y = np.array (df.iloc[4:, 2].tolist())
y_avg = []
for item in y:
    y_avg.append(float(item))
y = np.array (df.iloc[4:, 3].tolist())
y_max = []
for item in y:
    y_max.append(float(item))



fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Duration of execution (go function)")
ax.set_xlabel('Time')
ax.set_ylabel('Duration [ms]')
ax.plot(x, y_min, label = "minimum duration")
ax.plot(x, y_avg, label = "average duration")
ax.plot(x, y_max, label = "maximum duration")
ax.grid(True, linewidth=0.5)
plt.legend()
plt.show()
```

```python
#plotting the invocations

import matplotlib.pyplot as plt
import numpy as np

list1 = df.iloc[4:,0].tolist()

list2 = []
for x in list1:
    list2.append(x.split(" "))

list3 = []
for x in list2:
    list3.append(x[-1])

x = np.array(list3)


y = np.array (df.iloc[4:, 1].tolist())
invocations = []
for item in y:
    invocations.append(int(item))

fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Number of invocations per minute (go function)")
ax.set_xlabel('Time')
ax.set_ylabel('Number of invocations')
ax.plot(x, invocations)
ax.grid(True, linewidth=0.5)
plt.show()
```
